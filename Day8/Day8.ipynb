```python
import numpy as np
```

```python
with open('Day8Input.txt','r') as file:
    lines = [list(line.strip()) for line in file.readlines()]

grid=np.array(lines)
print(grid)

unique_elements = np.unique(grid)
print(unique_elements)

rowcount = grid.shape[0]
colcount = grid.shape[1]

def OutputAntiNodes (grid, node):
    indices = np.argwhere(grid==node)
    antinodes=[]
    #print(indices)
    for ind in indices:
        for ind2 in indices:
            if not np.all(ind == ind2):
                anti1 = [2*ind[0]-ind2[0],2*ind[1]-ind2[1]]
                anti2 = [2*ind2[0]-ind[0],2*ind2[1]-ind[1]]
                #print(anti1,anti2)
                if anti1[0]>=0 and anti1[0]<=rowcount-1 and anti1[1]>=0 and anti1[1]<=colcount-1: 
                    antinodes.append(anti1)
                if anti2[0]>=0 and anti2[0]<=rowcount-1 and anti2[1]>=0 and anti2[1]<=colcount-1: 
                    antinodes.append(anti2)

    antinodesunique=[]
    [antinodesunique.append(val) for val in antinodes if val not in antinodesunique]

    return antinodesunique

antinodesoutput = []
for item in unique_elements:
    if not item == '.':
        antinodesoutput.extend(OutputAntiNodes(grid,item))

antinodesoutputunique=[]
[antinodesoutputunique.append(val) for val in antinodesoutput if val not in antinodesoutputunique]

print(len(antinodesoutputunique),antinodesoutputunique)
```

```python
with open('Day8Input.txt','r') as file:
    lines = [list(line.strip()) for line in file.readlines()]

grid=np.array(lines)
print(grid)

unique_elements = np.unique(grid)
print(unique_elements)

rowcount = grid.shape[0]
colcount = grid.shape[1]

def hcf(int1,int2):
    #highest common factor
    if int1 == 0 or int2 == 0:
        return 1
    for i in range(min(int1,int2),0,-1):
        if int1 % i ==0 and int2 % i==0:
            return int(i)



def OutputNewAntiNodes (grid, node):
    indices = np.argwhere(grid==node)
    antinodes=[]
    #print(indices)
    for ind in indices:
        for ind2 in indices:
            if not np.all(ind == ind2):                
                row_gap = ind[0]-ind2[0]
                col_gap = ind[1]-ind2[1]
                #print('index1', ind, 'index2', ind2, 'row', row_gap,'col', col_gap)
                highestcommonfactor = hcf(abs(row_gap),abs(col_gap))
                row_gap = row_gap / highestcommonfactor
                col_gap = col_gap / highestcommonfactor
                
                j=0
                while ind[0]-j*row_gap in range(0, rowcount) and ind[1]-j*col_gap in range(0, colcount):
                    antinodes.append([ind[0]-j*row_gap,ind[1]-j*col_gap])
                    #print([ind[0]-j*row_gap,ind[1]-j*col_gap])
                    j+=1

                j=0
                while ind[0]+j*row_gap in range(0, rowcount) and ind[1]+j*col_gap in range(0, colcount):
                    antinodes.append([ind[0]+j*row_gap,ind[1]+j*col_gap])
                    #print([ind[0]+j*row_gap,ind[1]+j*col_gap])
                    j+=1

    antinodesunique=[]
    [antinodesunique.append(val) for val in antinodes if val not in antinodesunique]

    return antinodesunique

antinodesoutput = []
for item in unique_elements:
    if not item == '.':
        antinodesoutput.extend(OutputNewAntiNodes(grid,item))

antinodesoutputunique=[]
[antinodesoutputunique.append(val) for val in antinodesoutput if val not in antinodesoutputunique]

print(len(antinodesoutputunique),antinodesoutputunique)
```
