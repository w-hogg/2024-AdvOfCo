PART ONE

```python
with open('Day5Rules.txt', 'r') as file:
    rules = [rule.strip() for rule in file.readlines()]

def isAllowed (x,y):
    #is x and y allowed in the order shown? 
    #ie if y|x is in the list then give a false otherwise give a true
    return f"{y}|{x}" not in rules

def isUpdateAllowed(list):
    output = True
    for i in range(0, len(list)-1):
        x=list[i]
        for j in range (i, len(list)):
            y=list[j]
            if isAllowed(x,y) ==False:
                output = False
    return output      
```

```python
with open('Day5Updates.txt', 'r') as file:
    updates = [list(map (int,update.strip().split(','))) for update in file.readlines()]

AllowedUpdates=[isUpdateAllowed(update) for update in updates]

sum=0
for index, update in enumerate(updates):
    if AllowedUpdates[index]:
        sum += update[len(update) // 2]


sum
```

PART TWO

```python
import numpy as np
NotAllowedUpdates = [update for update , allowed in zip(updates,AllowedUpdates) if not allowed]
```

```python

def correct_order_update (update, rules_split):
    rules_to_keep = [rule for rule in rules_split if rule[0] in update and rule[1] in update]
    np_rules_to_keep = np.array(rules_to_keep)

    unique_elements_asc = np.unique(np_rules_to_keep[:,0], return_counts=True)
    sorted_unique_elements_asc = unique_elements_asc[0][np.argsort(-unique_elements_asc[1])]

    unique_elements_desc = np.unique(np_rules_to_keep[:,1], return_counts=True)
    sorted_unique_elements_desc = unique_elements_desc[0][np.argsort(unique_elements_desc[1])]

    ordered_update = np.append(sorted_unique_elements_asc,sorted_unique_elements_desc[0])
    return ordered_update.tolist()
```

```python
rules_split = [list(map (int,rule.split('|'))) for rule in rules]

Ordered_Updates = [correct_order_update(update, rules_split) for update in NotAllowedUpdates]
print(Ordered_Updates)

sum=0
for update in Ordered_Updates:
    sum += update[len(update) // 2]

print(sum)
```
