```python
Dict_Nums = {}
Dict_Rules = {}
Dict_Dependents ={}
Dict_Precedents ={}



def ImportToString (filepath):
    with open(filepath,'r') as file:
        Inputs = [line for line in file.read().split('\n\n')]
        Starting = [start for start in Inputs[0].split('\n')]
        Rules = [rule for rule in Inputs[1].split('\n')]
    return Starting, Rules

def FindDeps(in_con):
    output_set = set()
    rule = Dict_Rules[in_con]
    con_1, combine, con_2 = rule.split(' ')
    if con_1[0] == 'x' or con_1[0] == 'y':
        output_set.add(con_1)
    else:
        output_set = output_set.union(FindDeps(con_1))

    if con_2[0] == 'x' or con_2[0] == 'y':
        output_set.add(con_2)
    else:
        output_set = output_set.union(FindDeps(con_2))

    return output_set

def FindPrecs(in_con):
    output_set = set()
    for conn, rule in Dict_Rules.items():
        con_1, combine, con_2 = rule.split(' ')
        if con_1 == in_con or con_2 == in_con:
            if conn[0]=='z':
                output_set.add(conn)
            else:
                output_set = output_set.union(FindPrecs(conn))

    return output_set


Scenario ='Input'

Starting, Rules = ImportToString('Day24'+Scenario+'.txt')

for start_inital in Starting:
    Dict_Nums[start_inital[:3]] = int(start_inital[-1])

for rule in Rules:
    rule_set, rule_ans = rule.split(' -> ')
    Dict_Rules[rule_ans]=rule_set

for key in Dict_Rules.keys():
    Dict_Dependents[key]= FindDeps(key)

for key in Dict_Nums.keys():
    Dict_Precedents[key]= FindPrecs(key)

print(len(Dict_Rules),Dict_Rules)
print(len(Dict_Nums), Dict_Nums)
print(len(Dict_Dependents), Dict_Dependents)
print(len(Dict_Precedents), Dict_Precedents)

for r in range(0, 44):
    str_r = str(r)
    str_r = str_r.zfill(2)
    print('z'+str_r, len(Dict_Dependents['z'+str_r]), (r+1)*2, sorted(Dict_Dependents['z'+str_r]))

for r in range(0, 45):
    str_r = str(r)
    str_r = str_r.zfill(2)
    print('x'+str_r, len(Dict_Precedents['x'+str_r]), sorted(Dict_Precedents['x'+str_r]))
    print('y'+str_r, len(Dict_Precedents['y'+str_r]), sorted(Dict_Precedents['y'+str_r]))


```
