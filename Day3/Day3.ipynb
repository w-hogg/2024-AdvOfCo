```python
df2=df
firstmatch = [m.start(0) for m in re.finditer(pattern, file_contents)] 

df2['StartOfInstance']=firstmatch
df2.head()
```

```python
Dos = [m.start(0) for m in re.finditer(r'do\(\)', file_contents)] 
DoNots = [m.start(0) for m in re.finditer(r'don\'t\(\)', file_contents)] 
Dos
```

```python
def find_LargestInList_SmallerThanX(list,x): 
    pos = bisect.bisect_left(list, x) 
    return list[pos-1] if pos > 0 else None

def MostRecentDo(x):
    ans = find_LargestInList_SmallerThanX(Dos,x)
    return ans

def MostRecentDoNot(x):
    ans = find_LargestInList_SmallerThanX(DoNots,x)
    return ans
```

```python
df3 = df2
df3['NearestDo']=df3['StartOfInstance'].apply(MostRecentDo).fillna(99999999)
df3['NearestDoNot']=df3['StartOfInstance'].apply(MostRecentDoNot).fillna(0)
df3['CountOrNot'] = df3['NearestDo']>df3['NearestDoNot']
df3['MultiplyFinal'] = df3['CountOrNot']*df3['Multiply']
FinalSum = df3['MultiplyFinal'].sum()
df3
```
