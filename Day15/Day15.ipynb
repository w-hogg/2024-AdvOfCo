SETUP

```python
import numpy as np

def ImportToString (filepath):
    with open(filepath,'r') as file:
        line = file.read().replace('\n','')
    return line

def ImportToList (filepath):
    with open(filepath,'r') as file:
        line = file.read().replace('\n','')
        list = [item for item in line]
    return list

def ImportToArray(filepath):
    with open(filepath,'r') as file:
        lines = [list(line.strip()) for line in file.readlines()]
    return np.array(lines)

def ImportToListOfLists(filepath):
    with open(filepath,'r') as file:
        lines = [list(line.strip()) for line in file.readlines()]
    return lines

```

PART ONE

```python
Directions = {'^': [-1, 0], '<': [0,-1], '>': [0,1], 'v':[1,0]}

def Move(Grid, direction):
    SelectedDirection = Directions[direction]
    LocationPoint = np.argwhere(Grid=='@')[0]
    #print(LocationPoint)

    Search0Location = LocationPoint+SelectedDirection
    os_in_path = 0
    while Grid[Search0Location[0], Search0Location[1]] =='O':
        Search0Location=Search0Location+SelectedDirection
        os_in_path +=1

    if Grid[Search0Location[0], Search0Location[1]] == '.':
        #There is a space to move to 
        NewLocation = LocationPoint+SelectedDirection
        Grid[NewLocation[0], NewLocation[1]] = '@'
        Grid[LocationPoint[0], LocationPoint[1]] = '.'
        if os_in_path >0:
            Grid[Search0Location[0], Search0Location[1]] = 'O'

    return Grid

def MoveMany(Grid, directionstring):
    for item in directionstring:
        Grid = Move(Grid, item)
        #print(Grid)
    return Grid

def SumUpGrid(Grid):
    LocationsOfO = np.argwhere(Grid=='O')
    sum = 0
    for location in LocationsOfO:
        sum += location[0]*100 + location[1]
    return sum
    

Scenario = 'Input'

DirectionsImport = ImportToString('Day15'+Scenario+'Directions.txt')
Grid = ImportToArray('Day15'+Scenario+'Config.txt')

print(DirectionsImport)
print(Grid)

Grid = MoveMany(Grid, DirectionsImport )
print(Grid)
print(SumUpGrid(Grid))
```

```python
Directions = {'^': [-1, 0], '<': [0,-1], '>': [0,1], 'v':[1,0]}

def ExpandGrid(Grid):
    NewGrid = np.full((Grid.shape[0],Grid.shape[1] * 2), '.',dtype=Grid.dtype)
    for c in range(0,Grid.shape[1]):
        NewGrid[:,c*2] = np.where(Grid[:,c] == 'O', '[', Grid[:,c])
        NewGrid[:,c*2+1] = np.where(Grid[:,c] == 'O', ']', Grid[:,c])
        NewGrid[:,c*2+1] = np.where(NewGrid[:,c*2+1]=='@', '.',NewGrid[:,c*2+1])
    return NewGrid

def MoveNew(Grid, index, direction):
    SelectedDirection = Directions[direction]
    NewPlace = [x + y for x, y in zip(index, SelectedDirection)]

    if direction in ('<','>'):
        if Grid[NewPlace[0],NewPlace[1]] in ('[',']'):
            Grid = MoveNew(Grid, NewPlace, direction)
    else:
        if Grid[NewPlace[0],NewPlace[1]] in ('['):
            Grid = MoveNew(Grid, NewPlace, direction)    
            Grid = MoveNew(Grid, [NewPlace[0],NewPlace[1]+1], direction)
        elif Grid[NewPlace[0],NewPlace[1]] in (']'):
            Grid = MoveNew(Grid, NewPlace, direction)    
            Grid = MoveNew(Grid, [NewPlace[0],NewPlace[1]-1], direction)    
  
    Grid[NewPlace[0],NewPlace[1]] = Grid[index[0],index[1]]
    Grid[index[0],index[1]] = '.'

    return Grid


def CanItMove(Grid, index, direction):
    SelectedDirection = Directions[direction]
    NewPlace = [x + y for x, y in zip(index, SelectedDirection)]
    CanIt = True

    if direction in ('<','>'):
        if Grid[NewPlace[0],NewPlace[1]] in ('[',']'):
            CanIt = CanItMove(Grid, NewPlace, direction) and CanIt
        elif Grid[NewPlace[0],NewPlace[1]] == '#':
            CanIt = False
    else:
        if Grid[NewPlace[0],NewPlace[1]] in ('['):
            CanIt = CanItMove(Grid, NewPlace, direction) and CanIt
            CanIt = CanItMove(Grid, [NewPlace[0],NewPlace[1]+1], direction) and CanIt
        elif Grid[NewPlace[0],NewPlace[1]] in (']'):
            CanIt = CanItMove(Grid, NewPlace, direction) and CanIt
            CanIt = CanItMove(Grid, [NewPlace[0],NewPlace[1]-1], direction) and CanIt        
        elif Grid[NewPlace[0],NewPlace[1]] == '#':
            CanIt = False

    return CanIt


def MoveMany(Grid, directionstring):
    for item in directionstring:
        index = np.argwhere(Grid=='@')[0].tolist()
        #print(item)
        if CanItMove(Grid, index, item):
        #if True:
            Grid = MoveNew(Grid, index, item)
        #print(Grid)
    return Grid

def SumUpGrid(Grid):
    LocationsOfO = np.argwhere(Grid=='[').tolist()
    sum = 0
    for location in LocationsOfO:
        sum += location[0]*100 + location[1]
    return sum
    

Scenario = 'Input'

DirectionsImport = ImportToString('Day15'+Scenario+'Directions.txt')
Grid = ImportToArray('Day15'+Scenario+'Config.txt')
Grid = ExpandGrid(Grid)

print(DirectionsImport)
print(Grid)

Grid = MoveMany(Grid, DirectionsImport )
print(Grid)
print(SumUpGrid(Grid))
```
