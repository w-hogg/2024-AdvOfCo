```python
from itertools import permutations

DictNumercExclusions={('7','A'): 'vvv>>', ('4', 'A'): 'vv>>',('1', 'A'): 'v>>',
                      ('7','0'): 'vvv>', ('4', '0'): 'vv>',('1', '0'): 'v>',
                      ('A', '7'):'<<^^^',('A', '4'):'<<^^', ('A', '1'):'<<^', 
                      ('0', '7'):'<^^^',('0', '4'):'<^^', ('0', '1'):'<^' }

DictDirectionExclusions={('<','^'): '^>', ('<', 'A'): '^>>', 
                         ('^','<'): '<v', ('A', '<'): '<<v',}

DictMinLengts={}

def NumericRow(Digit):
    if Digit in ('7','8','9'):
        return 0 
    elif Digit in ('4','5','6'):
        return 1
    elif Digit in ('1','2','3'):
        return 2
    else: 
        return 3

def DirectionRow(Dir):
    if Dir in ('^','A'):
        return 0 
    else: 
        return 1

def NumericCol(Digit):
    if Digit in ('1','4','7'):
        return 0 
    elif Digit in ('0','2','5','8'):
        return 1
    else: 
        return 2

def DirectionCol(Dir):
    if Dir in ('<'):
        return 0 
    elif Dir in ('^','v'):
        return 1
    else: 
        return 2

def NumericPadOptions(From, To):
    ColShift = NumericCol(To)-NumericCol(From)
    RowShift = NumericRow(To)-NumericRow(From)

    Movement =''
    if ColShift>0:
        Movement += '>'*ColShift
    elif ColShift <0:
        Movement += '<'* -ColShift

    if RowShift>0:
        Movement += 'v'*RowShift
    elif RowShift <0:
        Movement += '^'*-RowShift

    options = set(permutations(Movement))
    MovementsSet = {''.join(option) for option in options }

    if (From, To) in DictNumercExclusions:
        MovementsSet.remove(DictNumercExclusions[(From,To)])

    return MovementsSet

def DirectionPadOptions(From, To):
    ColShift = DirectionCol(To)-DirectionCol(From)
    RowShift = DirectionRow(To)-DirectionRow(From)

    Movement =''
    if ColShift>0:
        Movement += '>'*ColShift
    elif ColShift <0:
        Movement += '<'* -ColShift

    if RowShift>0:
        Movement += 'v'*RowShift
    elif RowShift <0:
        Movement += '^'*-RowShift      
    
    options = set(permutations(Movement))
    MovementsSet = {''.join(option) for option in options }
    if (From, To) in DictDirectionExclusions:
        MovementsSet.remove(DictDirectionExclusions[(From,To)])
    return MovementsSet

def NumericStringOptions(NumericString):

    Steps = len(NumericString)
    NumericString = 'A'+ NumericString

    MovementsSet={''}
    
    for step in range(0, Steps):
        Options = NumericPadOptions(NumericString[step], NumericString[step+1])
        New_Set = set()
        for current in MovementsSet:
            for option in Options:
                New_Set.add(current + option+'A')
        
        MovementsSet = New_Set

        #print(MovementsSet)

    return MovementsSet

def DirectionPadExample(From, To):
    ColShift = DirectionCol(To)-DirectionCol(From)
    RowShift = DirectionRow(To)-DirectionRow(From)

    Movement =''
    if ColShift>0:
        Movement += '>'*ColShift
    elif ColShift <0:
        Movement += '<'* -ColShift

    if RowShift>0:
        Movement += 'v'*RowShift
    elif RowShift <0:
        Movement += '^'*-RowShift      

    return Movement

def DirectionStringExample(DirectionString):
    Steps = len(DirectionString)
    DirectionString = 'A'+DirectionString

    Movement=''
    for step in range(0, Steps):

        Movement += DirectionPadExample(DirectionString[step], DirectionString[step+1])
        Movement +='A'
    #print(Movement)
    return Movement

def DirectionStringOptions(DirectionString):
    Steps = len(DirectionString)
    DirectionString = 'A'+DirectionString

    MovementsSet={''}
    for step in range(0, Steps):
        Options = DirectionPadOptions(DirectionString[step], DirectionString[step+1])
        New_Set = set()
        for current in MovementsSet:
            for option in Options:
                New_Set.add(current + option+'A')
        
        MovementsSet = New_Set
        #print(MovementsSet)
    return MovementsSet

def ScoreString(NumericString):
    LengthOfPattern = MinLength(NumericString)
    StringInts = int(NumericString[:-1])
    print(LengthOfPattern, StringInts)
    return LengthOfPattern *StringInts

def MinLength(String):

    numeric_string_set = NumericStringOptions(String)
    Final_Set = set()
    for string_item in numeric_string_set:
        Final_Set.update(DirectionStringOptions(string_item))

    #print(Final_Set)

    Final_Set_2 = set()
    for string_item in Final_Set:
        Final_Set_2.add(DirectionStringExample(string_item))

    min_length = min(len(item) for item in Final_Set_2)

    return min_length

### PROGRAM ###
Strings=['029A', '980A', '179A', '456A', '379A'] #Test
#Strings =['083A','935A','964A', '149A', '789A'] #Input
#Strings=['029A']

totalscore = 0 
for NumericString in Strings:
    print(NumericString)
    totalscore += ScoreString(NumericString)

print(totalscore)


```
