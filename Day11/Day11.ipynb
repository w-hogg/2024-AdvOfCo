SETUP

```python
import numpy as np

def ImportToString (filepath):
    with open(filepath,'r') as file:
        line = file.read().replace('\n','')
    return line

def ImportToList (filepath):
    with open(filepath,'r') as file:
        line = file.read().replace('\n','')
        list = [item for item in line]
    return list

def ImportToListSpaces (filepath):
    with open(filepath,'r') as file:
        line = file.read().replace('\n','')
        list = [int(item) for item in line.split(' ')]
    return list

def ImportToArray(filepath):
    with open(filepath,'r') as file:
        lines = [list(line.strip()) for line in file.readlines()]
    return np.array(lines)

def ImportToListOfLists(filepath):
    with open(filepath,'r') as file:
        lines = [list(line.strip()) for line in file.readlines()]
    return lines
```

PART ONE

```python
Stones = ImportToListSpaces('Day11Test.txt')

blinks=25

print(Stones)

for blink in range(0, blinks):
    StonesNew=[]
    for stone in Stones:
        stone_length = len(str(stone))
        if stone == 0:
            StonesNew += [1]
        elif stone_length % 2 == 0:
            second_stone_string = str(stone)[int(stone_length/2):]
            first_stone_string = str(stone)[:int(stone_length/2)]
            StonesNew += [int(first_stone_string)]
            StonesNew += [int(second_stone_string)]
        else:
            StonesNew += [stone*2024]
    Stones = StonesNew

print(Stones)

print(len(Stones))


```

PART TWO

```python
Dict = {}

def CountStone(stepstone):
    step = stepstone[0]
    stone = stepstone[1]
    count = 0
    stone_length = len(str(stone))

    if (step, stone) in Dict:
        count = Dict[(step,stone)]
    else:
        if step == 0:
            count=1
        elif stone == 0:
            count += CountStone([step-1,1])
        elif stone_length % 2 == 0:
            second_stone_string = str(stone)[int(stone_length/2):]
            first_stone_string = str(stone)[:int(stone_length/2)]
            count += CountStone([step-1,int(first_stone_string)])
            count +=CountStone([step-1,int(second_stone_string)])
        else:
            count+= CountStone([step-1,stone*2024])

        Dict[(step, stone)] = count

    return count


# START OF CODE

blinks=75

Stones = ImportToListSpaces('Day11Input.txt')
StepStones = [[blinks,stone] for stone in Stones]
print(StepStones)

count = 0
for stepstone in StepStones:
    count += CountStone(stepstone)

#print(Dict)
print(count)
```
