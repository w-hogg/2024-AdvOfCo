```python
def ImportBlocks (filepath):
    with open(filepath,'r') as file:
        list = [block for block in file.read().split('\n\n')]
    return list

def Lock_to_Num(lock):
    block_nums = ['','','','','']
    block_list = lock.split('\n')

    for lock_row in range(1,7):
        for num, val in enumerate(block_list[lock_row]):
            if val == '.' and block_nums[num]=='':
                block_nums[num]=lock_row-1

    return block_nums

def Key_to_Num(key):
    block_nums = ['','','','','']
    block_list = key.split('\n')

    for key_row in range(6,-1,-1):
        for num, val in enumerate(block_list[key_row]):
            if val == '.' and block_nums[num]=='':
                block_nums[num]=5-key_row

    return block_nums




Scenario = 'Input'

Blocks = ImportBlocks('Day25'+Scenario+'.txt')

Keys=[]
Locks=[]
for block in Blocks:
    if block[0] =='#':
        Locks.append(block)
    else:
        Keys.append(block)


print(Blocks)
print(Keys, Locks)

Lock_Codes = []
for lock in Locks:
    Lock_Codes.append(Lock_to_Num(lock))

Key_Codes = []
for key in Keys:
    Key_Codes.append(Key_to_Num(key))


print(Lock_Codes)
print(Key_Codes)

total_count = 0
for lock in Lock_Codes:
    for key in Key_Codes:
        valid = True
        for pos in range(0,5):
            if lock[pos]+key[pos] > 5:
                valid = False
        if valid:
            total_count +=1

print(total_count)



```
