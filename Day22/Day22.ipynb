```python
Dict_of_Bananas = {}

def ImportToList (filepath):
    with open(filepath,'r') as file:
        line = file.readlines()
        list = [int(item.strip('\n')) for item in line]
    return list

def BitXoR(num_a,num_b):
    BinStr_a, BinStr_b = ConvertToBinary(num_a), ConvertToBinary(num_b)
    max_len = max (len(BinStr_a), len(BinStr_b))
    BinStr_a, BinStr_b = BinStr_a.zfill(max_len), BinStr_b.zfill(max_len)

    output = ''
    for bin_a, bin_b in zip(BinStr_a, BinStr_b):
        if bin_a == bin_b:
            output +='0'
        else: 
            output +='1'
    
    output_int = int(output, 2)

    return output_int

def ConvertToBinary(int_10):
    BinaryString = format(int_10,'b')
    return BinaryString

def NextSequence(num):
    num = BitXoR(num, 64*num) % 16777216
    num = BitXoR(num, num//32) % 16777216
    num = BitXoR(num, num*2048) % 16777216
    return num 

def ManySequences(num, sequence_count):
    num_list = [num]
    for number in range(0, sequence_count):
        num = NextSequence(num)
        num_list.append(num)

    One_Digit = [item %10 for item in num_list]
    
    One_Digit_Diff = []
    for indic in range(0, len(One_Digit) - 1):
        One_Digit_Diff.append(One_Digit[indic+1]-One_Digit[indic])
    
    return One_Digit, One_Digit_Diff

def find_series(lst, series):
    series_length = len(series)
    for i in range(len(lst) - series_length + 1):
        if lst[i:i + series_length] == series:
            return i
    return -1


Scenario ='Input'
Max_Secrets = 2000
NumList = ImportToList('Day22'+Scenario+'.txt')
#Secret_Num = 123



counting = 0

for secret in NumList:
    One_Digit_List, One_Digit_Diff = ManySequences(secret, Max_Secrets)
    #print(len(One_Digit_List), One_Digit_List)
    #print(len(One_Digit_Diff), One_Digit_Diff)
    counting +=1
    if counting%100 ==0:
        print(counting)

    New_Dict = {}
    for int_find in range(Max_Secrets-4, -1,-1):
        series = One_Digit_Diff[int_find: int_find+4]
        Num_Bananas = One_Digit_List[int_find+4]
        New_Dict[tuple(series)] = Num_Bananas
        #print(series, Num_Bananas)

    for item in New_Dict:
        if item in Dict_of_Bananas:
            Dict_of_Bananas[item] += New_Dict[item]
        else:
            Dict_of_Bananas[item] = New_Dict[item]



print(Dict_of_Bananas)

#print(Dict_of_Bananas[(-2,1,-1,3)])

max_bananas = max(banana for banana in Dict_of_Bananas.values())

print(max_bananas)
```
