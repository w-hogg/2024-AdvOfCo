```python
def action_old(array):
    indices = np.where(array == 'u')
    end = False
    if indices[0].size > 0:
        # action an up
        if indices[0] == 0:
            array[indices[0],indices[1]] = '1'
            end = True
        elif array[indices[0]-1,indices[1]]== '.' or array[indices[0]-1,indices[1]]== '1':
            array[indices[0]-1,indices[1]] = 'u'
            array[indices[0],indices[1]] = '1'
        elif array[indices[0]-1,indices[1]]== '#':
            array[indices[0],indices[1]] = 'r'
        return array, end

    indices = np.where(array == 'r')
    if indices[0].size > 0:
        if indices[1] == array.shape[1]-1:
            array[indices[0],indices[1]] = '1'
            end = True
        # action a down
        elif array[indices[0],indices[1]+1]== '.' or array[indices[0],indices[1]+1]== '1':
            array[indices[0],indices[1]+1] = 'r'
            array[indices[0],indices[1]] = '1'
        elif array[indices[0],indices[1]+1]== '#':
            array[indices[0],indices[1]] = 'd'
        return array, end

    indices = np.where(array == 'd')
    if indices[0].size > 0:
        # action a down
        if indices[0] == array.shape[0]-1:
            array[indices[0],indices[1]] = '1'
            end = True
        elif array[indices[0]+1,indices[1]]== '.' or array[indices[0]+1,indices[1]]== '1':
            array[indices[0]+1,indices[1]] = 'd'
            array[indices[0],indices[1]] = '1'
        elif array[indices[0]+1,indices[1]]== '#':
            array[indices[0],indices[1]] = 'l'
        return array, end

    indices = np.where(array == 'l')
    if indices[0].size > 0:
        # action a down
        if indices[1] == 0:
            array[indices[0],indices[1]] = '1'
            end = True
        elif array[indices[0],indices[1]-1]== '.' or array[indices[0],indices[1]-1]== '1':
            array[indices[0],indices[1]-1] = 'l'
            array[indices[0],indices[1]] = '1'
        elif array[indices[0],indices[1]-1]== '#':
            array[indices[0],indices[1]] = 'u'
        return array, end

```

```python
def action(array):
    directions = {'u': (-1, 0, 'r'), 'r': (0, 1, 'd'), 'd': (1, 0, 'l'), 'l': (0, -1, 'u')}
    end = False

    for direction, (di, dj, next_direction) in directions.items():
        indices = np.where(array == direction)
        if indices[0].size > 0:
            i, j = indices[0][0], indices[1][0]
            ni, nj = i + di, j + dj

            if (direction == 'u' and i == 0) or (direction == 'r' and j == array.shape[1] - 1) or \
               (direction == 'd' and i == array.shape[0] - 1) or (direction == 'l' and j == 0):
                array[i, j] = '1'
                end = True
            elif array[ni, nj] in ('.', '1'):
                array[ni, nj] = direction
                array[i, j] = '1'
            elif array[ni, nj] == '#':
                array[i, j] = next_direction

            return array, end
    return array, end
```

```python
# Read the content of the file
with open('Day6Input.txt', 'r') as file:
    lines = file.readlines()
    grid = np.array([list(line.strip()) for line in lines])

grid[np.where(grid == '^')]='u'

#print(grid)
end = False
loop = False
count = 0
while not end :
    grid, end = action(grid)
    count += 1
    #print(grid)

count_ones = np.count_nonzero(grid == ('1'))
print(count)
print(count_ones)
```

```python
def IsLoopedGrid(grid):
    
    grid[np.where(grid == '#')] = 0
    i,j = np.where(grid == '^')
    grid[i,j]='.'
    h=0

    while h==0:

        while grid[i-1,j] == '.':
            i=i-1
            if i==0:
                return False
        grid[i-1,j] = int(grid[i-1,j]) + 1
        if grid[i-1,j]==5:
            return True
        
        grid[i,j]='^'
        print(grid)
        grid[i,j]='.'

        while grid[i,j+1] == '.':
            j=j+1
            if j+1 == grid.shape[1]-1:
                return False
        grid[i,j+1] = int(grid[i,j+1]) + 1
        if grid[i,j+1]==5:
            return True
        
        grid[i,j]='^'
        print(grid)
        grid[i,j]='.'

        while grid[i+1,j] == '.':
            i=i+1
            if i+1 == grid.shape[0]-1:
                return False
        grid[i+1,j] = int(grid[i+1,j]) + 1
        if grid[i+1,j]==5:
            return True

        grid[i,j]='^'
        print(grid)
        grid[i,j]='.'

        while grid[i,j-1] == '.':
            j=j-1
            if j==0:
                return False
        grid[i,j-1] = int(grid[i,j-1]) + 1
        if grid[i,j-1]==5:
            return True

        grid[i,j]='^'
        print(grid)
        grid[i,j]='.'
        
```
