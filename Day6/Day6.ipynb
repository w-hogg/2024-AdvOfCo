```python
def action_old(array):
    indices = np.where(array == 'u')
    end = False
    if indices[0].size > 0:
        # action an up
        if indices[0] == 0:
            array[indices[0],indices[1]] = '1'
            end = True
        elif array[indices[0]-1,indices[1]]== '.' or array[indices[0]-1,indices[1]]== '1':
            array[indices[0]-1,indices[1]] = 'u'
            array[indices[0],indices[1]] = '1'
        elif array[indices[0]-1,indices[1]]== '#':
            array[indices[0],indices[1]] = 'r'
        return array, end

    indices = np.where(array == 'r')
    if indices[0].size > 0:
        if indices[1] == array.shape[1]-1:
            array[indices[0],indices[1]] = '1'
            end = True
        # action a down
        elif array[indices[0],indices[1]+1]== '.' or array[indices[0],indices[1]+1]== '1':
            array[indices[0],indices[1]+1] = 'r'
            array[indices[0],indices[1]] = '1'
        elif array[indices[0],indices[1]+1]== '#':
            array[indices[0],indices[1]] = 'd'
        return array, end

    indices = np.where(array == 'd')
    if indices[0].size > 0:
        # action a down
        if indices[0] == array.shape[0]-1:
            array[indices[0],indices[1]] = '1'
            end = True
        elif array[indices[0]+1,indices[1]]== '.' or array[indices[0]+1,indices[1]]== '1':
            array[indices[0]+1,indices[1]] = 'd'
            array[indices[0],indices[1]] = '1'
        elif array[indices[0]+1,indices[1]]== '#':
            array[indices[0],indices[1]] = 'l'
        return array, end

    indices = np.where(array == 'l')
    if indices[0].size > 0:
        # action a down
        if indices[1] == 0:
            array[indices[0],indices[1]] = '1'
            end = True
        elif array[indices[0],indices[1]-1]== '.' or array[indices[0],indices[1]-1]== '1':
            array[indices[0],indices[1]-1] = 'l'
            array[indices[0],indices[1]] = '1'
        elif array[indices[0],indices[1]-1]== '#':
            array[indices[0],indices[1]] = 'u'
        return array, end

```

```python
def action(array):
    directions = {'u': (-1, 0, 'r'), 'r': (0, 1, 'd'), 'd': (1, 0, 'l'), 'l': (0, -1, 'u')}
    end = False

    for direction, (di, dj, next_direction) in directions.items():
        indices = np.where(array == direction)
        if indices[0].size > 0:
            i, j = indices[0][0], indices[1][0]
            ni, nj = i + di, j + dj

            if (direction == 'u' and i == 0) or (direction == 'r' and j == array.shape[1] - 1) or \
               (direction == 'd' and i == array.shape[0] - 1) or (direction == 'l' and j == 0):
                array[i, j] = '1'
                end = True
            elif array[ni, nj] in ('.', '1'):
                array[ni, nj] = direction
                array[i, j] = '1'
            elif array[ni, nj] == '#':
                array[i, j] = next_direction

            return array, end
    return array, end
```
