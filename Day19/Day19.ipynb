PART ONE

```python
def ImportToList (filepath):
    with open(filepath,'r') as file:
        line = file.read().replace('\n','')
        list = [item for item in line.split(', ')]
    return list

def ImportToListByLine(filepath):
    with open(filepath,'r') as file:
        lines = [line.strip() for line in file.readlines()]
    return lines

def CreateADesign(InputList, Patterns):
    fulldesign = ''
    for item in InputList:
        fulldesign += Patterns[item]

    return fulldesign

def IsThisPossible(Patterns, FinalDesign):

    DesignCode = [0]
    Pattern_count = len(Patterns)

    while True:
        design = CreateADesign(DesignCode, Patterns)
        if design == FinalDesign: #if design is Final Design then return True
            #print('True', design, DesignCode)
            return True
        
        if design == FinalDesign[:len(design)]: #if design matches FinalDesign for first digits then add a new DesignCode and ReSet
            DesignCode.append(0)

        else: #if it doesn't work then add one more
            DesignCode[-1] += 1
            while DesignCode[-1] == Pattern_count: #remove Codes that have reached the end
                if DesignCode[0] == Pattern_count:
                    #print('False', design)
                    return False
                DesignCode.pop()
                DesignCode[-1] +=1

def IsThisPossibleCount(Patterns, FinalDesign):

    DesignCode = [0]
    Pattern_count = len(Patterns)
    number_of_ways =0

    while True:
        design = CreateADesign(DesignCode, Patterns)
        #print(DesignCode)
        if design == FinalDesign: #if design is Final Design then return True
            print('True', number_of_ways, design, DesignCode)
            number_of_ways +=1

        if design == FinalDesign[:len(design)] and not (design == FinalDesign): #if design matches FinalDesign for first digits then add a new DesignCode and ReSet
            DesignCode.append(0)

        else: #if it doesn't work then add one more
            DesignCode[-1] += 1
            while DesignCode[-1] == Pattern_count: #remove Codes that have reached the end
                if DesignCode[0] == Pattern_count:
                    print(number_of_ways, FinalDesign)
                    return number_of_ways
                DesignCode.pop()
                DesignCode[-1] +=1            

Scenario = 'Test'

Patterns = ImportToList('Day19'+Scenario+'Patterns.txt')
Designs = ImportToListByLine('Day19'+Scenario+'Designs.txt')

print(len(Patterns),Patterns)
print(len(Designs),Designs)

count = 0
for design in Designs:
    count += IsThisPossible(Patterns, design)

print('Possible Designs', count)

count = 0
for design in Designs:
    count += IsThisPossibleCount(Patterns, design)

print('Number of ways', count)
```

```python
def ImportToList (filepath):
    with open(filepath,'r') as file:
        line = file.read().replace('\n','')
        list = [item for item in line.split(', ')]
    return list

def ImportToListByLine(filepath):
    with open(filepath,'r') as file:
        lines = [line.strip() for line in file.readlines()]
    return lines


DictPossibilities = {}

def CountPossibilities(design):
    global Patterns
    count=0

    if design in DictPossibilities:
        return DictPossibilities[design] 
    
    for pattern in Patterns:
        if pattern == design:
            count += 1
        elif design[:len(pattern)]==pattern:
            count += CountPossibilities(design[len(pattern):])

    DictPossibilities[design] = count

    return count        

Scenario = 'Input'

Patterns = ImportToList('Day19'+Scenario+'Patterns.txt')
Designs = ImportToListByLine('Day19'+Scenario+'Designs.txt')

print(len(Patterns),Patterns)
print(len(Designs),Designs)

count = 0
for design in Designs:
    count += CountPossibilities(design)

print('Number of ways', count)
```
