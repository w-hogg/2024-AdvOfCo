```python
import re

filepath = 'Day13Input.txt'    
with open(filepath,'r') as file:
    line = file.read()

pattern = re.compile(r'Button A: X\+(\d+), Y\+(\d+)\s*Button B: X\+(\d+), Y\+(\d+)\s*Prize: X=(\d+), Y=(\d+)')
matches = pattern.findall(line)

extracted_numbers_only = []

for match in matches:
    button_a_x, button_a_y, button_b_x, button_b_y, prize_x, prize_y = map(int, match)
    extracted_numbers_only.append([
        button_a_x, button_a_y,
        button_b_x, button_b_y,
        prize_x,  prize_y]
    )

# print(extracted_numbers_only)

totalsum = 0
for entry in extracted_numbers_only:
    print(entry)
    a_presses, b_presses = 0, 0
    button_a_x, button_a_y, button_b_x, button_b_y, prize_x, prize_y = entry
    method = ''
    
    if not button_a_x*button_b_y == button_a_y*button_b_x:
        method = 'simultaneous eqs'
        b_presses = (prize_y*button_a_x - prize_x*button_a_y) / (button_a_x*button_b_y - button_a_y*button_b_x)
        a_presses = (prize_y*button_b_x - prize_x*button_b_y) / (button_a_y*button_b_x- button_a_x*button_b_y)

        if not(int(a_presses) == a_presses and int(b_presses) == b_presses):
            a_presses, b_presses = 0,0                 
        if a_presses > 100 or b_presses >100:
            a_presses, b_presses = 0,0

    else:
        if button_a_x > 3* button_b_x:
            method = 'Equivalent: start with a'
            a_presses, b_presses = 0, 0
            sum_x, sum_y =0,0
            while prize_x >= sum_x+button_a_x and a_presses<=99:
                sum_x += button_a_x
                sum_y += button_a_y
                a_presses +=1

            while prize_x >= sum_x+button_b_x and b_presses<=99:
                sum_x += button_b_x
                sum_y += button_b_y
                b_presses +=1

            if not(sum_x == prize_x and sum_y == prize_y):
                a_presses, b_presses = 0
            

        else:
            method = 'Equivalent: start with a'
            a_presses, b_presses = 0, 0
            sum_x, sum_y =0,0

            while prize_x >= sum_x+button_b_x and b_presses<=99:
                sum_x += button_b_x
                sum_y += button_b_y
                b_presses +=1

            while prize_x >= sum_x+button_a_x and a_presses<=99:
                sum_x += button_a_x
                sum_y += button_a_y
                a_presses +=1

            if not(sum_x == prize_x and sum_y == prize_y):
                a_presses, b_presses = 0

    
    totalsum += 3*a_presses + b_presses
    
    if a_presses == 0 and b_presses ==0:
        print (f'N: Method: {method}')
    else:
        print (f'Y: A presses: {a_presses}, B presses: {b_presses}, Method: {method}')

print(totalsum)



```
